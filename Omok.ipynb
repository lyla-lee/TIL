```
import numpy as np
def vis_board(board):
	h, w = board.shape

	s = " "
	for x in range(w):
		s += "{:2d}".format(x % 10)
	print(s)
	for y in range(h):
		s = "{:2d}".format(y % 100)
		for x in range(w):
			if board[y, x] == 0: 
				if x == 0 and y == 0: s += chr(0x250C)
				elif x == 0 and y == h - 1:	s += chr(0x2514)
				elif x == w - 1 and y == 0: s += chr(0x2510)
				elif x == w - 1 and y == h - 1: s += chr(0x2518)
				elif x == 0: s += chr(0x251C)
				elif x == w - 1: s += chr(0x2524)
				elif y == 0: s += chr(0x252C)
				elif y == h - 1: s += chr(0x2534)
				else: s += chr(0x253C)
			elif board[y, x] == 1: s += chr(0x25CB)
			elif board[y, x] == 2: s += chr(0x25CF)
		print(s)
    
```

```
def check(board,x,y,flag):
    n=0
    min_i = min(4,x)
    max_i = min(5, w-x)
    for i in range(-min_i,max_i):
        if board[y,x+i] != flag: n=0
        else:
            n+=1
            if n ==5: return True
    n = 0
    min_i=min(4,y)
    max_i=min(5,h-y)
    for i in range(-min_i,max_i):
        if board[y+i,x] != flag: n = 0
        else:
            n +=1
            if n==5: return True
    n=0
    min_i = min(4,x,y)
    max_i = min(5,w-x, h-y)
    for i in range(-min_i,max_i):
        if board[y+i,x+i] != flag: n=0
        else:
            n+=1
            if n==5: return True
    n=0
    min_i = min(4,x,h-y-1)
    max_i = min(5,w-x,y+1)
    for i in range(-min_i,max_i):
        if board[y-i,x+i] != flag: n = 0
        else:
            n += 1
            if n ==5:return True
    return False
```

```
if __name__=='__main__':
    print('input board size...')
    h,w=list(map(int,input().split()))
    board=np.zeros((h,w))
    vis_board(board)
    flag=1
    
    while True:
        if flag==1: print('White')
        else: print('Black')
        step=input().split()
        if step==['0']:
            print('game over')
            break
        
        x,y=list(map(int,step))
        if x>=w and y>=h or board[y,x] != 0:
            print('cannot put here')
            continue
        board[y,x]=flag
        vis_board(board)
        
        if check(board,x,y,flag):
            if flag==1:
                print('White win')
                break
            else:
                print('Black win')
                break
        flag=2/flag
        
```
